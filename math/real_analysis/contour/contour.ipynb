```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def f(x, y):
    return x * np.exp(-(x**2) - y**2) + y / 6
```

### HÃ¶henlinien

```python
n = 500
x = np.linspace(-2.1, 2.1, n)
y = np.linspace(-2.1, 2.1, n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

cs = plt.contour(X, Y, Z, cmap="viridis")
plt.clabel(cs)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label="z")
plt.show()
```

```python
n = 500
x = np.linspace(-2.1, 2.1, n)
y = np.linspace(-2.1, 2.1, n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, levels=1000, cmap="viridis")
plt.colorbar(label="z")
plt.xlabel("x")
plt.ylabel("y")
cs = plt.contour(X, Y, Z, levels=10, colors="white", alpha=0.7)
plt.clabel(cs)
plt.show()
```

### Isokonturlinien (Partielle Funktionen)

```python
levels = [-2, -1, 0, 1, 2]
n = 500
x = np.linspace(-2.1, 2.1, n)
y = np.linspace(-2.1, 2.1, n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

cs = plt.contour(Y, Z, X, levels=levels)
plt.clabel(cs)
plt.xlabel("y")
plt.ylabel("z")
plt.colorbar(label="x")
plt.grid()
plt.show()
```
