```python
import numpy as np
from matplotlib import pyplot as plt
```

### Gradient and Contourf

```python
def f(x, y):
    return y * np.exp(-(x**2))


n = 200
x = np.linspace(-2, 2, n)
y = np.linspace(-2, 2, n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contourf(X, Y, Z, cmap="viridis")
plt.colorbar(label="f(x,y)")

n2 = 15
x2 = np.linspace(-2, 2, n2)
y2 = np.linspace(-2, 2, n2)
X2, Y2 = np.meshgrid(x2, y2)
Z2 = f(X2, Y2)
grady, gradx = np.gradient(Z2, y2, x2)
plt.quiver(X2, Y2, gradx, grady, angles="xy")

plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
