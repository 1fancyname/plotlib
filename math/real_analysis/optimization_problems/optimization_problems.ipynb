```python
import numpy as np
from matplotlib import pyplot as plt
```

2. $\displaystyle g(x,y) = \frac{1}{4}x^2 + \frac{1}{2}y^2 - \frac{1}{4}xy$

    - Kritische Punkte:

        $$ (x,y) = (0,0) $$

    - Hesse-Matrix:

        $$
        \mathbf{H}_g(x,y) = \begin{pmatrix} \tfrac{1}{2} & -\tfrac{1}{4} \\ -\tfrac{1}{4} & 1 \end{pmatrix} = \mathbf{H}_g(0,0)
        $$

    - Eigenwerte:

        $$
        \det\begin{pmatrix}
        \frac{1}{2}-\lambda &- \frac{1}{4}\\
        - \frac{1}{4} & 1 - \lambda\\
        \end{pmatrix} = \left( \tfrac{1}{2} - \lambda \right) \cdot \left( 1 - \lambda \right) - \left(- \tfrac{1}{4} \right)^2 = \lambda^2 - \tfrac{3}{2}\lambda  +\tfrac{7}{16} = 0
        $$

        $$
        \Rightarrow\quad \lambda_\pm = \frac{3}{4} \pm \frac{1}{2\sqrt{2}}\quad\Rightarrow\quad \lambda_+ \approx 1.10 \quad\text{und}\quad \lambda_- \approx 0.396
        $$

        Zwei positive Eigenwerte, also ist die Hesse-Matrix positiv definit und es liegt ein lokales Minimum vor.


```python
n = 200
xmin, xmax = -2, 2
ymin, ymax = -2, 2
x = np.linspace(xmin, xmax, n)
y = np.linspace(ymin, ymax, n)
X, Y = np.meshgrid(x, y)

# contour of function
Z = 0.25 * X**2 + 0.5 * Y**2 - 0.25 * X * Y
cs = plt.contour(X, Y, Z)
plt.clabel(cs, inline=1, fontsize=10)
plt.colorbar(label="g(x,y)")
# plot minimum
plt.plot(0, 0, "ro", label="Minimum")

# plot gradient in a 10x10 grid
grad_y, grad_x = np.gradient(Z, x, y)
# plt.streamplot(X, Y, grad_x, grad_y)
s = n // 10
plt.quiver(
    X[::s, ::s],
    Y[::s, ::s],
    grad_x[::s, ::s],
    grad_y[::s, ::s],
    angles="xy",
    scale_units="xy",
    label="Gradient"
)

# Eigenvektoren der Hesse-Matrix darstellen
H = np.array([[1 / 2, -1 / 4], [-1 / 4, 1]])
ew, ev = np.linalg.eig(H)
plt.quiver(
    [0, 0],
    [0, 0],
    ev[0] * ew,
    ev[1] * ew,
    color="m",
    angles="xy",
    scale_units="xy",
    scale=1,
    label="Eigenvector",
)

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.title("Eigenvectors of Hessematrix (scaled with eigenvalues)")
plt.show()
```

3. $\displaystyle h(x,y) = 4xy-9+2x^2+8y-2y^2+4x $

    - Kritische Punkte:

        $$
        (x,y) = \left(-\tfrac{3}{2}, \tfrac{1}{2}\right)
        $$

    - Hesse-Matrix:

        $$
        \mathbf{H}_h(x,y) = \begin{pmatrix} 4 & 4 \\ 4 & -4 \end{pmatrix} = \mathbf{H}_h\left(-\tfrac{3}{2}, \tfrac{1}{2}\right)
        $$

    - Eigenwerte:

        $$
        \det\begin{pmatrix} 4-\lambda & 4 \\ 4 & -4-\lambda \end{pmatrix} = \left(4-\lambda\right)\cdot \left(-4-\lambda\right) - 4^2 = \lambda^2 - 32 = 0
        $$

        $$
        \Rightarrow\quad \lambda_\pm = \pm 4\sqrt{2}
        $$

        Die Hesse-Matrix ist indefinit, da sie einen positiven und einen negativen Eigenwert hat. Der Punkt $(x,y) = \left(-\tfrac{3}{2}, \tfrac{1}{2}\right)$ ist also ein Sattelpunkt.


```python
# Hhenlinien und kritischer Punkt skizzieren
Z = 4 * X * Y - 9 + 2 * X**2 + 8 * Y - 2 * Y**2 + 4 * X
cs = plt.contour(X, Y, Z)
plt.clabel(cs, inline=1, fontsize=10)
plt.colorbar(label="h(x,y)")
plt.plot(-3 / 2, 1 / 2, "bo", label="Stationary Point")

# Eigenvektoren der Hesse-Matrix
ew, ev = np.linalg.eig([[4, 4], [4, -4]])
plt.quiver(
    [-3 / 2, -3 / 2],
    [1 / 2, 1 / 2],
    ev[0] * ew,
    ev[1] * ew,
    color="m",
    angles="xy",
    scale_units="xy",
    scale=5,
    label="Eigenvectors",
)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
